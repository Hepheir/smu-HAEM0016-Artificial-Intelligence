# 1. 집값 데이터 읽기

```python
import numpy as np
import pandas as pd

data = pd.read_csv('../data/house_price_us.txt', names=['size', 'n_room', 'price'])
print(data)

X = np.c_[data['size'], data['n_room']]
y = data['price']
m = len(data)
```

```python
X = np.array(X).reshape(m, 2)
y = np.array(y).reshape(m, 1)

print(X.shape, y.shape)
```

# 2. 그래프 그리기

```python
import matplotlib.pyplot as plt

plt.plot(X[:,0].reshape(-1), y, 'b.')
plt.xlabel('Size of a house in ft^2')
plt.ylabel('Price in USD')
plt.show()

plt.plot(X[:,1].reshape(-1), y, 'b.')
plt.xlabel('# of rooms')
plt.ylabel('Price in USD')
plt.show()
```

# 3. Feature scaling(normalization)

$$
x_i = \frac{x_i - \mu_i}{s_i}
$$

```python
def featureNormalize(X):
    mu = np.mean(X, axis=0)
    std = np.std(X, axis=0)

    X_norm = (X-mu) / std

    return X_norm, mu, std
```

```python
X, mu, std = featureNormalize(X)
print(X.shape)
```

# 4. Gradient descent -> theta

```python
X_b = np.c_[np.ones((m, 1)), X]

learning_rate = 0.01
n_iter = 400

theta = np.random.randn(3,1)
gradients = np.zeros((3,1))

for i in range(n_iter):
    gradients = 2.0/m * X_b.T.dot(X_b.dot(theta)-y)
    theta -= learning_rate * gradients

print('theta:')
print(theta)
```

# 5. 임의의 집 크기(x_my)에 대한 집값 예측하기

```python
X_mine = np.array([[890, 2]])
X_mine = (X_mine - mu) / std
print(X_mine)

X_mine_b = np.c_[np.ones((1,1)), X_mine]
y_predict = X_mine_b.dot(theta)
print(y_predict, "달러")
```

# 6. 반복 횟수에 따른 비용함수 $J(\theta)$ 그리기

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ComputeJ(t0, t1, t2, X, y):
    m = len(X)
    X_b = np.c_[np.ones((m, 1)), X]
    theta = np.array([t0, t1, t2])

    cost_vector = X_b.dot(theta) - y
    cost_vector = cost_vector.reshape(-1)
    cost2 = np.square(cost_vector)
    cost_sum = np.sum(cost2) / (2*m)

    return cost_sum
```

```python
X = np.c_[data['size'], data['n_room']]
y = data['price']
m = len(data)

X = np.array(X).reshape(m, 2)
y = np.array(y).reshape(m, 1)
print(X.shape, y.shape)

X, mu, std = featureNormalize(X)
print(X.shape)
```

```python
X_b = np.c_[np.ones((m, 1)), X]

learning_rate = 0.01
n_iter = 400

theta = np.random.randn(3,1)
gradients = np.zeros((3, 1))

itr_arr = []
cost_arr = []
for i in range(n_iter):
    gradients = 2.0/m * X_b.T.dot(X_b.dot(theta) - y)
    theta -= learning_rate * gradients

    cost = ComputeJ(theta[0], theta[1], theta[2], X, y)
    cost_arr.append(cost)
    itr_arr.append(i)

print('theta:')
print(theta)
```

```python
plt.plot(itr_arr, cost_arr)
plt.xlabel("number of iterations")
plt.ylabel('cost J')
plt.show()
```

## 6-1. learning rate를 바꿔가면서 비교

```python
learning_rate_list = (0.0001, 0.001, 0.01, 0.1)

for k, learning_rate in enumerate(learning_rate_list):
    n_iter = 400

    theta = np.random.randn(3,1)
    gradients = np.zeros((3, 1))

    itr_arr = []
    cost_arr = []
    for i in range(n_iter):
        gradients = 2.0/m * X_b.T.dot(X_b.dot(theta) - y)
        theta -= learning_rate * gradients

        cost = ComputeJ(theta[0], theta[1], theta[2], X, y)
        cost_arr.append(cost)
        itr_arr.append(i)

    plt.subplot(1, len(learning_rate_list), k+1)
    plt.plot(itr_arr, cost_arr)
    plt.xlabel("number of iterations")
    plt.ylabel('cost J')
    plt.title(f'lr={learning_rate}')

plt.show()
```
